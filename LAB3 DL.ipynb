```python
"""
=============================================================
Deep Learning Lab Series — LAB 3
Course: Deep Learning (22AIE304) | Batch: 2024-2028
NAME: PREM SANTH C K
ROLL NUMBER: CH.SC.U4AIE24056
=============================================================
"""

import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.datasets import mnist
from sklearn.model_selection import ParameterGrid

def build_advanced_mlp(hidden_units=64, activation='relu', dropout_rate=0.0):
    model = models.Sequential([
        layers.Dense(hidden_units, activation=activation, input_shape=(2,)),
        layers.Dropout(dropout_rate),
        layers.Dense(hidden_units, activation=activation),
        layers.Dropout(dropout_rate),
        layers.Dense(1, activation='sigmoid')
    ])
    return model


def run_lab3():
    print("=" * 60)
    print("LAB 3: Hyperparameter Optimization (Grid Search)")
    print("=" * 60)

    # Using XOR for quick demonstration
    X = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=np.float32)
    y = np.array([0, 1, 1, 0], dtype=np.float32)

    param_grid = {
        'learning_rate': [0.01, 0.1],
        'hidden_units':  [8, 16],
        'dropout_rate':  [0.0, 0.2],
    }
    # Limit to 5 trials for lab
    grid = list(ParameterGrid(param_grid))[:5]

    results = []
    for i, params in enumerate(grid, 1):
        model = build_advanced_mlp(
            hidden_units=params['hidden_units'],
            activation='relu',
            dropout_rate=params['dropout_rate']
        )
        model.compile(optimizer=optimizers.Adam(learning_rate=params['learning_rate']),
                      loss='binary_crossentropy', metrics=['accuracy'])
        history = model.fit(X, y, epochs=100, verbose=0, validation_data=(X, y))
        train_loss = history.history['loss'][-1]
        val_acc    = history.history['val_accuracy'][-1]
        overfit    = "No" if (val_acc >= 0.9) else "Yes"
        results.append((i, params, train_loss, val_acc, overfit))
        print(f"Trial {i} | {params} | Loss={train_loss:.4f} | Val Acc={val_acc:.2f} | Overfit={overfit}")

    best = max(results, key=lambda r: r[3])
    print(f"\nBest config: {best[1]} → Val Acc={best[3]:.2f}")
    return results

if __name__ == "__main__":
    lab3_results = run_lab3()
```
